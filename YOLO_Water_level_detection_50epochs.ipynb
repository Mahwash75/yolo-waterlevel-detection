```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import os


def train_yolo(data_yaml, model='yolov8n.pt', epochs=50, img_size=640, save_dir='runs/train'):
    """
    Train the YOLO model.
    :param data_yaml: Path to the YAML configuration file for the dataset.
    :param model: Pretrained YOLO model to use (e.g., yolov8n.pt, yolov8m.pt).
    :param epochs: Number of training epochs.
    :param img_size: Size of input images.
    :param save_dir: Directory to save training results.
    """
    # Initialize the YOLO model
    yolo_model = YOLO(model)

    # Train the model
    yolo_model.train(data=data_yaml, epochs=epochs, imgsz=img_size, project=save_dir)
    print(f"Model training completed. Results saved in {save_dir}")

    return yolo_model


def display_predictions(model, test_images_dir, num_examples=5, conf=0.25):
    """
    Display predictions on test images with bounding boxes and class labels.
    :param model: Trained YOLO model.
    :param test_images_dir: Path to the directory containing test images.
    :param num_examples: Number of examples to display.
    :param conf: Confidence threshold for predictions.
    """
    test_images = [os.path.join(test_images_dir, img) for img in os.listdir(test_images_dir) if img.endswith(('.jpg', '.png'))]
    test_images = test_images[:num_examples]

    for img_path in test_images:
        # Read the image
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Make predictions
        results = model.predict(img_path, conf=conf, imgsz=640)
        boxes = results[0].boxes  # Extract bounding boxes

        if len(boxes) == 0:
            print(f"No predictions for {os.path.basename(img_path)}")
            continue

        for box in boxes:
            # Extract box coordinates, confidence, and class label
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Convert to integers
            conf = float(box.conf[0])  # Confidence score
            cls = int(box.cls[0])  # Class index
            class_label = results[0].names[cls]  # Class name

            # Draw the bounding box
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

            # Annotate with the class label and confidence
            label = f"{class_label} {conf:.2f}"
            cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # Display the image
        plt.figure(figsize=(8, 8))
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Predictions for {os.path.basename(img_path)}")
        plt.show()



# Paths and configuration
data_yaml = "/content/drive/MyDrive/Yolo_glass_dataset/YOLO_glass_dataset.yaml"  # YAML file containing dataset info (classes, paths)
test_images_dir = "/content/drive/MyDrive/Yolo_glass_dataset/Yolo_glass_dataset/test/images"  # Path to the test images directory
model_path = "yolov8n.pt"  # Choose YOLO model: yolov8n.pt, yolov8m.pt, etc.
save_directory = '/content/drive/MyDrive/Yolo_glass_dataset/Train_YOLO_model'

# Train the model
trained_model = train_yolo(data_yaml=data_yaml, model=model_path, epochs=50, img_size=640, save_dir=save_directory)

# Display predictions on test examples
display_predictions(trained_model, test_images_dir, num_examples=5, conf=0.25)
```
